#### What is the SQLite Database

SQLite is a C-language library that implements a small, fast, self-contained, high-reliability, full-featured, SQL database engine.

SQLite is the most used database engine in the world. SQLite is built into all mobile phones and most computers and comes bundled inside countless other applications that people use every day. More Information...

```python
import sqlite3 as sqlt

#Connecting with the databases . It will automatically create the db if it is not present.
db = sqlt.connect('FileIO and DB Connection Examples\\my_database4.db')

# Execute method is used to excute the statement passes as parameter.
db.execute("drop table if exists test1")

# To create a multipline statement use the triple quotes.
query = """ 
CREATE TABLE TEST1( a varchar(20),b varchar(20),c REAL, d INTEGER);
"""

db1 = sqlt.connect('FileIO and DB Connection Examples\\mydata2.sqlite')
db1.execute(query)
db1.commit()
```

```python
 datas = [('Atlanta','Georgia',1.25,6),
         ('Tallahassee','Florida',2.6,3),
         ('Sacramento','California',1.7,5)]

stmt = "INSERT INTO test1 VALUES (?,?,?,?)"

db1.executemany(stmt,datas)
db1.commit()
```
